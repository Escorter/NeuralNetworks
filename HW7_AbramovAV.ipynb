### Архитектура YOLOv5

   YOLOv5 относится к архитектуре One-Stage detector - подход, который предсказывает координаты определённого количества bounding box'ов с результатами классификации и вероятности нахождения объекта, и в дальнейшем корректируя их местоположение. На самом деле YOLOv5 это тот же самый YOLOv4, только реализованный на Pytorch.
   
   Вообще, YOLO это современная архитектура нейронных сетей, которая предназначена для распознавания или обнаружения объектов на изображении. YOLO полезна, когда возникает необходимость быстро сделать кастомную модель для детекции, не вникая в дебри компьютерного зрения и многообразие моделей.
   
   В привычном для нас понимании в распознавании объектов изображение всегда разбивается на некоторые области, а после происходит их классификация. При таком подходе мы просматриваем изображение дважды: в первый раз мы находим области, которые содержат объект, во второй раз мы классифицируем. А вот архитектура YOLO просматривает изображение единожды, во время которого изображение сжимается до квадратной матрицы, в каждой из элементов которой записана информация о содержании искомого объекта и его классе.
   Архитектуру YOLOv5 можно представить в следующем виде:
  ![title](OneStageDetector.jpeg)
  
  Сеть сжимает исходное изображение в несколько feature map'ов с использованием skip-connection и прочих архитектурных фишек. Полученные карты признаков приводятся в одно разрешение с помощью апсемплинга и конкатенируются. Затем предсказываются классы и bounding box'ы для объектов, далее для каждого объекта выбирается самый вероятный bounding box с помощью Non-Maximum Suppression.
  
  Из достоинств данной архитектуры необходимо отметить, что сеть начинает выдавать хоть и не безупречный, но приемлемый результат даже уже после сотни обучающих сэмплов, и далее, по мере роста обученности сети ее качество неуклонно растет.
